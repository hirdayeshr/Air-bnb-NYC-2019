```python
data.fillna({"reviews_per_month":0},inplace=True)
data.fillna({"last_review":0},inplace=True)
data.fillna({"host_name ":0},inplace=True)
data=data[data['reviews_per_month']!="NaN"]
```

```python
names = []
for i in data.name:
    names.append(i)
    
def splitname(name):
    s = str(name).split()
    return s
nameswbw =[]

for x in names:
    for j in splitname(x):
        j = j.lower()
        nameswbw.append(j)
```
